# 1. 从fundemantals.csv开始！

fundemantals.csv 是这些股票的年报数据

请用数据回答以下问题：

1. S&P500股票在2015年`net income`的均值是多少？最大值比最小值多多少？（每问10分，共计20分）
2. S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是多少？固定资产占总资产比例最小的股票的代码（ticker symbol）是什么？（每问10分，共计20分）


```python
fd = pd.read_csv('fundamentals.csv')
fd.head()
```

```python
#第一题
fun2015 = fd.query('`For Year` == 2015.0')
print('S&P500股票在2015年net income的均值是',fun2015['Net Income'].mean())
print('最大值比最小值多',fun2015['Net Income'].max() - fun2015['Net Income'].min())
print()

#第二题
fun2016 = fd.query('`For Year` == 2016.0')
funRatio = fun2016.rename(columns={'Fixed Assets':'Fixed_Assets', 'Total Assets':'Total_Assets'})#命名不规范，新手两行泪
funRatio['Fixed/Total'] = (funRatio.Fixed_Assets/funRatio.Total_Assets)

print('S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是',funRatio['Fixed/Total'].mean())
print('固定资产占总资产比例最小的股票的代码（ticker symbol）是',funRatio.groupby(['Fixed/Total'])['Ticker Symbol'].min().iat[0])
```

# 2. 加入securities.csv~

securities.csv包含了这些股票的基本信息

1. 请列举出各个sector中的加入时间最早的股票名称（10分）
2. 请列举出每一个州中加入时间最晚的股票名称（10分）


```python
se = pd.read_csv('securities.csv')
se.head(n = 10)
```

```python
se = se.rename(columns={'Date first added':'Date_first_added',
                        'GICS Sector':'GICS_Sector',
                        'Address of Headquarters':'Address_of_Headquarters'}) #遇事不决先改名
se['Date_first_added'] = se['Date_first_added'].fillna('null')
indexOfNull = se[(se.Date_first_added=='null')].index.tolist()
se2 = se.drop(indexOfNull) #删除空值
se2.sort_values(by = 'Date_first_added')

#第一题
s1 = se2.sort_values(by = 'Date_first_added') #先按照日期排序
s1 = s1.set_index(keys =['GICS_Sector'])

x = [1,2,3,6]
s1 = s1.drop(s1.columns[x],axis = 1) #删掉多余列
sector = []
symbol = []
date = []
for i in range(0,len(s1.index),1):
    if not any(s1.index[i] in s for s in sector): #遇到新的sector，直接把第一个当做最早的
        sector.append(s1.index[i])
        String = s1.index[i] + ': ' + s1.iat[i,0]
        symbol.append(String)
        date.append(s1.iat[i,2])
    else: #判断加入时间是否相同
        for j,x in enumerate(sector):
            if s1.index[i] == x and s1.iat[i,2] == date[j]: #并列第一
                String = s1.index[i] + ': ' + s1.iat[i,0]
                symbol.append(String)
                break

print('各个sector中的加入时间最早的股票:')
for i,x in enumerate(symbol):
    print(x)

#第二题
s2 = se2.sort_values(by = 'Date_first_added',ascending=False) #先反过来排序
s2 = s2.set_index(keys=['Address_of_Headquarters'])
x2 = [1,2,3,4,6]
s2 = s2.drop(s2.columns[x2],axis = 1)#删掉多余列

address = []
symbol2 = []
date2 = []
for i in range(0,len(s2.index),1):
    if not any(s2.index[i] in s for s in address): #遇到新的州，直接把第一个当最晚的
        address.append(s2.index[i])
        String2 = s2.index[i] + ': ' + s2.iat[i,0]
        symbol2.append(String2)
        date2.append(s2.iat[i,1])
    else:#判断加入时间是否相同
        for j,x in enumerate(address):
            if s2.index[i] == x and s2.iat[i,1] == date2[j]:#并列倒一
                String2 = s2.index[i] + ': ' + s2.iat[i,0]
                symbol2.append(String2)
                break

print('\n每一个州中加入时间最晚的股票:')
for i,x in enumerate(symbol2):
    print(x)
```

1. 请思考，合并两个表的信息的时候，我们应该用什么样的准则对其它们（10分）
2. 请列举每个sector在2013-2016年累计Research&Development的总投入（10分）
3. 请列举出每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值（20分）

```python
df = pd.merge(fd, se, how = 'inner',on = 'Ticker Symbol')
df
```

```python
#第一题
df1 = df.query('`For Year` > 2012 and `For Year` < 2017') #筛选出2013-2016年的数据
Sector_sum = df1.groupby('GICS_Sector')['Research and Development'].sum() #按sector求和
print('每个sector在2013-2016年累计Research&Development的总投入:\n',Sector_sum)


#第二题
df2 = df1.set_index(keys=['GICS_Sector'])
col_n = ['Security','For Year','Research and Development']
df2 = pd.DataFrame(df2,columns = col_n) #刚发现的去掉多余列的方法……

sector2 = {} #key=sector, value=number of company
company = {} #key=company, value=sector
number = {} #key=company, value=number
least = {} #key=sector, value=company
com = ''#临时变量
for i in range(0,len(df2.index),1):
    
    #公司数不够3个
    if df2.index[i] not in sector2.keys(): # 新的sector
        sector2[df2.index[i]] = 1
        com = df2.iat[i,0]
        company[com] = df2.index[i]
        number[com] = float(df2.iat[i,2])
        least[df2.index[i]] = com   
    elif sector2[df2.index[i]] < 3 and df2.iat[i,0] == com and (i+1 == len(df2.index) or df2.iat[i+1,0] == com): #累加年份
        number[df2.iat[i,0]] +=  float(df2.iat[i,2])   
    elif sector2[df2.index[i]] < 3 and df2.iat[i,0] == com and (i+1 == len(df2.index) or not(df2.iat[i+1,0] == com)): #设置最菜公司
        number[df2.iat[i,0]] +=  float(df2.iat[i,2])
        if number[df2.iat[i,0]] < number[least[df2.index[i]]]:
            least[df2.index[i]] = com
    elif sector2[df2.index[i]] < 3: #同sector新公司
        sector2[df2.index[i]] += 1
        com = df2.iat[i,0]
        company[com] = df2.index[i]
        number[com] = float(df2.iat[i,2])
    
    #公司数多于3个
    elif sector2[df2.index[i]] >= 3 and not(df2.iat[i,0] in number.keys()): #新公司
        com = df2.iat[i,0]
        number[com] = float(df2.iat[i,2])
    elif sector2[df2.index[i]] >= 3 and df2.iat[i,0] == com and (i+1 == len(df2.index) or df2.iat[i+1,0] == com):#累加年份
         number[df2.iat[i,0]] +=  float(df2.iat[i,2])
    elif sector2[df2.index[i]] >= 3 and df2.iat[i,0] == com and (i+1 == len(df2.index) or not(df2.iat[i+1,0] == com)): #与最菜比较并更新最菜
        number[df2.iat[i,0]] +=  float(df2.iat[i,2])
        if number[df2.iat[i,0]] > number[least[df2.index[i]]]:
            del company[least[df2.index[i]]]
            company[com] = df2.index[i]
            tem_least = com
            for k,v in company.items(): #寻找下一个最菜
                if v == df2.index[i] and number[k] < number[tem_least]:
                    tem_least = k
            least[df2.index[i]] = tem_least #将临时最菜当做真的最菜

print('\n每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值:')
for k,v in sector2.items():
    print('Sector: ',k)
    for ke,va in company.items():
        if va == k:
            print('company: ',ke,'投入的数值: ',number[ke])
    print()
```

假设你是某基金公司的老板，现在对于每只股票，你都专门安排了一位负责它的交易员。公司规定每一位交易员手中的资金要么全部买入要么全部卖出（空仓，转化为现金）。假设2016年每一位交易员手中都有10000美元，假设他们都能够看到2016年全年的数据，假设他们都能抓住每一次机会，那么请问2016年底时，赚钱最多的股票是哪一只，赚了多少钱？

```python
p = pd.read_csv('prices.csv')
p
```

```python
col = ['date','symbol','open']
p1 = pd.DataFrame(p,columns = col) #删掉多于列

symbol = {}
com2 = ''
last_low = 0
for i in range(0, len(p1.index),1):
    if not p1.iat[i,1] in symbol.keys(): #遇到新的股票
        com2 = p1.iat[i,1]
        symbol[com2] = 0
        last_low = float(p1.iat[i,2])
    elif float(p1.iat[i,2]) > last_low: #发现上升序列，当场计算
        symbol[com2] = float(p1.iat[i,2]) - last_low
    elif float(p1.iat[i,2]) <= last_low: #上升中断，起点前移
        last_low = float(p1.iat[i,2])

max_num = 0
max_symbol = ''
for k,v in symbol.items(): #取出赚钱最多的股票
    if v >= max_num:
        max_symbol = k
        max_num = v
print('赚钱最多的股票是: ',max_symbol,',赚了: ',max_num*10000)
```
